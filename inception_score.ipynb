## Inception Score

Input - Path of folder with the image files that have been generated by the GAN.

Function - Takes the folder of images, resizes all of the images within the folder, calculates the Inception Score using the implementation found in the Salimans et al. paper. 

Output - Average and Standard Deviation of the Inception Score for each image, and hence the GAN image files.

```python
#Change image paths as necessary to the folder in which the images are located. 
image_list = glob.glob('/content/drive/MyDrive/BMENE4460/dataset/pix2pix_images_updated/*/*.png')
image_list_2 = glob.glob('/content/drive/MyDrive/BMENE4460/yutaro_testing/image_sort/*/*.png')

image_lists = [image_list, image_list_2]
```

```python
#For best usage, resize each images to (299,299,3). 
images = []
for image_list in image_lists:
  for file in image_list:
      image = Image.open(file)
      image = np.array(image)

      img_1 = cv2.resize(image[:,:,0], dsize=(299,299), interpolation=cv2.INTER_AREA)
      img_2 = cv2.resize(image[:,:,1], dsize=(299,299), interpolation=cv2.INTER_AREA)
      img_3 = cv2.resize(image[:,:,2], dsize=(299,299), interpolation=cv2.INTER_AREA)
      resized_img = np.array([img_1, img_2, img_3])
      images.append(resized_img)

images = np.array(images)

print(images.shape)
```

```python
inception_score_mean, inception_score_std = get_inception_score(images, splits = 1)
```
